```python
eleme1 = eleme1.drop("Unnamed: 24",axis = 1)
eleme1
```

```python
eleme1["团队id"] = eleme1["团队id"].astype("object")
eleme1["代理商id"] = eleme1["代理商id"].astype("object")
eleme1["是否达标"] = eleme1["代理商id"].astype("category")
```

#### 探索数据

```python
# 设置图表颜色
sns.set_palette("pastel")
# 设置图表尺寸
plt.rcParams["figure.figsize"] = [7.00, 3.50]
plt.rcParams["figure.autolayout"] = True
# 设置字体为黑体
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
```

##### 达标比例

```python
figure, axes = plt.subplots(1, 2)
sns.histplot(eleme1, x='综合得分', ax=axes[0])
sns.boxplot(eleme1, y='综合得分', ax=axes[1])
plt.show()
```

##### 得分呈现左态分布，60以下为极端情况，均分80分

```python
sns.scatterplot(eleme1, x='KPI得分', y='综合得分')
plt.show()
```

##### kpi对综合得分影响最大，除极个别情况都集中在中线上
